```python
import numpy as np  #importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

```python
df=pd.read_csv('regression_train.csv') # reading data
df.shape  # checking shape of data
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.dropna(axis=0, how='any', inplace=False)  #dropping null value
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
data3=pd.get_dummies(df['data3'],drop_first=True) # encoding data as we need all the values in intiger format 
data10=pd.get_dummies(df['data10'],drop_first=True) 
data9=pd.get_dummies(df['data9'],drop_first=True) 
```

```python
df=df.drop('data3',axis=1)# dropping the column
df=df.drop('data10',axis=1)
df=df.drop('data9',axis=1)
```

```python
df=pd.concat([df,data3],axis=1)# concating the columns to main data set
df=pd.concat([df,data10],axis=1)
df=pd.concat([df,data9],axis=1)
```

```python
df.dtypes
```

```python
df.corr()
```

```python
x=df
y=df.iloc[:,-9]
x
```

```python
y
```

```python
x=x.drop('output',axis=1)
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
multiple=LinearRegression()
multiple.fit(x_train,y_train)
```

```python
y_pred=multiple.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
```
